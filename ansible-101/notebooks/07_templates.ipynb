Another way of using templates is to enforcing policies, like:

  - logging centralization, naming  and rotating
  - limits enforcement
  - whatever
 
 
#### Log rotation
```
#
# Rotate {{programname}} logs generated by rsyslog_simple.j2 .
#
{{ "/".join(("/var/log/", dirname, "*.log")).replace("//", "/") }}  {
        missingok
        compress
        copytruncate
        daily
        rotate 31
        minsize 2048
        notifempty
}
```
